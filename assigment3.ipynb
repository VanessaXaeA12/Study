```python
#Libraries used

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Import files 

name = "mcdonalds_menu.csv"

# import menu and display the first two rows of the dataframe
df = pd.read_csv(name)
df.reset_index(inplace=True) #resetting the index (otherwise category is the new index and it messes up with merge)
```

```python
#The code 

#1. make new data frame with only those columns
df = df.loc[0:len(df),['Calories', 'Total Fat', 'Saturated Fat', 'Cholesterol', 'Sodium', 'Carbohydrates', 'Sugars', 'Protein']]

#2. plot data frame
pd.plotting.scatter_matrix(df, diagonal='kde', figsize=(20,12), grid=True)
```
